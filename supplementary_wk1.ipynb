# Pandas를 이용한 데이터 정리

````python
import pandas as pd
````

````python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
````

````python
df.describe()
````

````python
df["species"] = iris.target
df.head()
````

**`df.head()`**
- 데이터의 처음 5개 행을 출력하는 함수
- 데이터가 잘 들어갔는지 확인할 때 사용

````python
pd.plotting.scatter_matrix(df.iloc[:,:-1], figsize=(10, 10), c=df.species)
````

**산점도 행렬(scatter matrix)**<br>
- 각 특성(feature) 간의 관계를 시각화하는 산점도 행렬을 그림
- `pd.plotting.scatter_matrix()`는 모든 특성 조합에 대해 산점도를 그리는 함수
- `df.iloc[:, :-1]` : `iloc`을 사용해 마지막 열(species)을 제외한 모든 특성 데이터만 선택
- `figsize=(10, 10)` : 그림 크기를 10 x 10 인치로 설정
- `c=df.species` : `species` 값을 색깔로 사용하여, 다른 붓꽃 품종(0, 1, 2)에 따라 색을 다르게 표시

````python
from sklearn.preprocessing import LabelEncoder
````

**LabelEncoder**<br>
- 머신러닝 모델은 일반적으로 **숫자 데이터**를 다루기 때문에, **문자형(텍스트) 데이터를 숫자로 변환**해야 함
- `LabelEncoder`는 **텍스트(카테고리형 데이터)를 숫자로 변환하는 도구**

````python
le = LabelEncoder()
le.fit(iris.target_names)
````

**`le.fit(iris.target_names)`**
- LabelEncoder가 각 품종을 숫자로 변환할 준비를 함


```
{
    "setosa": 0,
    "versicolor": 1,
    "virginica": 2
}

```



````python
le.classes_
````

````python
le.transform(["setosa", "virginica"])
````

````python
le.inverse_transform([1, 2, 0])
````

````python
iris.target
````

````python
le.inverse_transform(iris.target)
````

# Softmax Regression Model 학습 및 평가

````python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class="multinomial", max_iter=5000, penalty=None) # Softmax regression
# model = LogisticRegression(multi_class="multinomial", max_iter=5000, penalty="l1", C=1.0, solver="saga") # with L1 regularization
# model = LogisticRegression(multi_class="multinomial", max_iter=5000, penalty="l2", C=1.0, solver="saga") # with L2 regularization
model.fit(iris.data, iris.target)

# target = "species"
# model.fit(df.drop(target, axis=1).values, df[target].values)
# target col drop
````

**LogisticRegression**<br>
- 이진 분류(Binary Classification) 및 다중 클래스 분류(Multiclass Classification) 문제에서 사용됨
- Softmax Regression(다중 클래스 로지스틱 회귀)로 사용 가능
- 데이터의 특성(Feature)와 목표값(Target)간의 관계를 선형 모델로 학습하여 클래스를 예측<br><br>

`multi_class="multinomial"`: 다중 클래스 분류(Softmax Regression) 사용<br>
`max_iter=5000`: 최대 5000번 반복하여 학습<br>
`penalty=None`: 정규화(Regularization) 미적용

````python
y_prob = model.predict_proba(iris.data)
y_prob[:10]
````

**`predict_proba()`**<br>
- 각 클래스에 속할 확률을 예측
- 각 행의 합 = 1

````python
y_cls = model.predict(iris.data)
y_cls[:10]
````

**`predict()`**<br>
각 샘플을 가장 확률이 높은 클래스로 예측

````python
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(iris.target, y_cls))
````

````python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
````

# Diabetes 데이터를 활용한 회귀 모형 학습/평가

````python
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
print(diabetes.DESCR)
````

````python
diabetes.keys()
````

````python
diabetes.feature_names
````

````python
diabetes.data[:5]
# age, sex 값이 저런 걸 보면 이미 datasetdl normalized된 걸 알 수 있음
# 아마 0.05, -0.04가 남자/여자일 것임(순서는 모름)
````

````python
diabetes.target[:10]
# lower할수록 안좋음
````

````python
from sklearn.linear_model import LinearRegression, Ridge, Lasso

model = LinearRegression()
# model = Ridge(alpha=1.0) # Ridge regression # 여기서 알파는 람다값! (라쏘도 동일)
# model = Lasso(alpha=1.0) # LASSO
model.fit(diabetes.data, diabetes.target)
````

````python
y_pred = model.predict(diabetes.data) # classification 아니니까 proba 없음
y_pred[:10]
````

````python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(diabetes.target, y_pred)
print("MSE:", mse)
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.plot(diabetes.target, y_pred, "o")

plt.title('True vs. Predicted Values')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')

# Optional: plot a 45-degree line for reference: 45도 라인은 y=y hat을 의미함 (y=x)
max_val = max(diabetes.target.max(), y_pred.max()) # Find the maximum value between true and predicted values... y max
min_val = min(diabetes.target.min(), y_pred.min()) # Find the minimum value between true and predicted values... y min
plt.plot([min_val, max_val], [min_val, max_val], '--', color='red', linewidth=2)
# 최소, 최댓값 사이에 라인 그림

plt.grid()
plt.show()
````

````python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
````
